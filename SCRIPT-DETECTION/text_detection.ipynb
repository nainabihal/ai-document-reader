```python
from paddleocr import PaddleOCR
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
detector = PaddleOCR(
    det=True,
    rec=False,
    cls=False,
    show_log=False
)
```

```python
ocr_en = PaddleOCR(lang="en", det=False, cls=False, show_log=False)
ocr_hi = PaddleOCR(lang="hi", det=False, cls=False, show_log=False)
```

```python
REC_MAP = {
    "en": ocr_en,
    "hi": ocr_hi
}
```

```python
det_result = detector.ocr("172589759_3_pg6.png")
```

```python
np.array(det_result[0][2][0]).astype(int)
```

```python
x, y, w, h = cv2.boundingRect(np.array(det_result[0][6][0]).astype(int))
```

```python
img = cv2.imread("172589759_3_pg6.png")
for line in range(len(det_result[0])):
    x, y, w, h = cv2.boundingRect(np.array(det_result[0][line][0]).astype(int))
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
```

```python
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
result = ocr_en.ocr(img, det=False, cls=False),ocr_hi.ocr(img, det=False, cls=False)
```

```python
text, conf = result[0][0][0]
```

```python
text=""
confidence=0
for out in result:
    if out[0] is not None:
        text, conf = out[0][0]
        if conf > confidence:
            text = text
            confidence = conf
```
